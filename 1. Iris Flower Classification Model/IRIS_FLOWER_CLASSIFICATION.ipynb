##### IMPORTING ALL THE LIBRARIES

```python
#importing all essential libraries of our project

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

##### LOADING THE DATASET

```python
#Loading the data

column = ['Sepal Lenght' , 'Sepal Width' , 'Petal Length' , 'Petal Width', 'Class_Label']
df = pd.read_csv('D:\\Data Science Projects\\LetsGrowMore Internship\\1. Iris Flower Classification Model\\iris.data', names=column)
df.head(50)
```

##### VISUALIZATION OF DATASET

```python
df.describe()
```

```python
#visualize the whole dataset

sns.pairplot(df, hue='Class_Label')
```

##### SEPARATE INPUT AND OUTPUT COLUMN

```python
#separate features and target 

data = df.values
X = data[:,0:4]
Y = data[:,4]
```

```python
#calculate average of each features for all clases

Y_Data = np.array([np.average(X[:,i][Y==j].astype('float32')) for i in range (X.shape[1]) for j in (np.unique(Y))])
Y_Data_reshaped = Y_Data.reshape(4,3)
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0,1)
X_axis = np.arange(len(column)-1)
width=0.25
```

```python
# plot the average

plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Verginica')
plt.xticks(X_axis, column[:4])
plt.xlabel("Features")
plt.ylabel("Values in cm.")
plt.legend(bbox_to_anchor=(1.3,1))
plt.show
```

##### SPLITTING THE DATA INTO TRAINNING AND TESTING

```python
# split the data to train and test dataset

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
print(X_test, Y_test)
```

##### MODEL-1: SUPPORT VECTOR MACHINE ALGORITHM

```python
#SUPPORT VECTOR MACHINE ALGORITHM

from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

##### SVC()

```python
prediction1 = model_svc.predict(X_test)

# calculate the accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction1)*100)
```

##### MODEL-2: LOGISTIC REGRESSION

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train, Y_train)
```

```python
prediction2 = model_LR.predict(X_test)

#calculate the accuracy

from sklearn.metrics import  accuracy_score
print(accuracy_score(Y_test, prediction2)*100)
```
